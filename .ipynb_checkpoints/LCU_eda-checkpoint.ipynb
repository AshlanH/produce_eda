## Initialization Phase + Data Info

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_rows', 500)
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('df_filtered.csv', index_col=0)
df_LCU = df[df.Element == 'Producer Price (LCU/tonne)'].reset_index(drop=True)
df_USD = df[df.Element == 'Producer Price (USD/tonne)'].reset_index(drop=True)
df_PPI = df[df.Element == 'Producer Price Index (2014-2016 = 100)'].reset_index(drop=True)
all_df = [df_LCU, df_USD, df_PPI]
```

### AVERAGE PRODUCE INDEX FOR EACH PRODUCT

```python
produce_index = df_PPI.groupby('Item').mean().sort_values(by='Value').reset_index(level=0)
produce_index_dictionary = {}
for index, row in produce_index.iterrows():
    produce_index_dictionary[row.Item] = row.Value
```

### Distribution of Produce by Year

```python
g = sns.catplot(
    data = df_USD,
    y = 'Item',
    x = 'Year',
    #aspect = 2.5,
    height = 20,
    kind='box',
    color = '#8cb8b7'
)

g.ax.set_title('Distribution of produce by year\n', fontdict={'fontsize':30})
g.ax.set_yticklabels(df_USD.Item.unique(), fontsize = '15')
g.ax.set_ylabel('Produce', fontsize = '21')
g.ax.set_xticks(df_USD.Year.unique())
g.ax.set_xticklabels(df_USD.Year.unique(), rotation = 90)
plt.show()
```


### Distribution of Produce Price by Year

```python
g = sns.catplot(
    data = df_USD,
    y = 'Item',
    x = 'Value',
    aspect = 1,
    height = 15,
    kind='box',
    color = '#8cb8b7'
)

g.ax.set_title('Distribution of produce price: Wide Ranges indicate volatile changes\n', fontdict={'fontsize':20})
g.ax.set_yticklabels(df_USD.Item.unique(), fontsize = '15')
g.ax.set_ylabel("Products",fontsize = '21')
#g.set(xlim=(0, 3000))
plt.show()
```

## Plotting Yearly Values to Yearly Growth Indexes

```python
def collect_data_field(item, start_date, end_date):
    a = df_PPI[(df_PPI.Item == item) & (df_PPI.Year >= start_date) & (df_PPI.Year <= end_date)].reset_index(drop=True)
    b = df_USD[(df_USD.Item == item) & (df_USD.Year >= start_date) & (df_USD.Year <= end_date)].reset_index(drop=True)
    #return pd.concat([a.iloc[:][['Item', 'Year', 'Value', 'Element']], b.iloc[:][['Item', 'Year', 'Value', 'Element']]], axis=1, join='inner')
    #return a.iloc[:][['Item', 'Year', 'Value', 'Element']].join(b.iloc[:][['Item', 'Year', 'Value', 'Element']], lsuffix = 'Item')
    b['Growth Rate'] = a.iloc[:]['Value']
    return b.iloc[:][['Item', 'Year', 'Value', 'Growth Rate']]
```

```python
df2 = collect_data_field('Bananas', 2000, 2013)
df2.head()
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
sns.lineplot(data = df2, x = 'Year', y = 'Value', ax = ax)
sns.barplot(data = df2, x = 'Year', y = 'Growth Rate', ax = ax, color = 'gray')
plt.show()
```
