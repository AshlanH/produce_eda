## Initialization Phase + Data Info

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_rows', 500)
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('malaysia_produce.csv')
df_LCU, df_SLC, df_USD, df_PPI = [df[df.Element == 'Producer Price (LCU/tonne)'], 
df[df.Element == 'Producer Price (SLC/tonne)'],
df[df.Element == 'Producer Price (USD/tonne)'], df[df.Element == 'Producer Price Index (2014-2016 = 100)']]
```

```python
# Extracting a segment of df_LCU data categorized by Item
df3 = df_LCU[df_LCU['Item'].isin(df_LCU.Item.unique()[8:14])].reset_index()
```

```python
df3.head()
```

```python
g = sns.relplot(data = df3[df3.Item == 'Chillies and peppers, green'], x = 'Year', y = 'Value',kind = 'line', aspect = 2)
```

```python
g = sns.catplot(data = df3, x='Value', y = 'Item',kind='box', aspect = 2.5, width = 0.7)
#g.set(xlim=(0, 10000))
g.ax.set_xticks([i for i in range (0, 10000, 1000)])
g.set_xticklabels([i for i in range (0, 10000, 1000)], rotation = 35)
plt.show()
```

```python
g = sns.relplot(
    data=df3,
    x = 'Year',
    y = 'Value',
    lw = 2,
    hue = 'Item',
    palette= 'magma',
    kind = 'line',
    height = 6,
    aspect = 2,
)
#g.ax.xlim(2000,2010)
g.ax.set_xticks([i for i in range (1990, 2020)])
g.set_xticklabels([i for i in range (1990, 2020)], rotation = 35)
#g.set(xlim=(2010,2020), ylim=(0, 5000))
#g.ax.set_xlabels([i for i in range (1990, 2020)], rotation=90)
plt.show()
```

```python
g = sns.lmplot(
    data=df3,
    x = 'Year',
    y = 'Value',
    #lw = 2,
    hue = 'Item',
    palette= 'magma',
    #kind = 'line',
    height = 6,
    aspect = 2,
)
#g.ax.xlim(2000,2010)
g.ax.set_xticks([i for i in range (1990, 2020)])
g.set_xticklabels([i for i in range (1990, 2020)], rotation = 35)
g.set(xlim=(2010,2020))
#g.ax.set_xlabels([i for i in range (1990, 2020)], rotation=90)
plt.show()
```
